```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Na = 6.023e23 # Avogadro's number
m = 1.007 / Na * 1e-3 # mass
kB = 1.3806e-23 # Boltzmann constant
h = 1.0546e-34 # Planck constant
L = 1e-9 # box size
```

```python
temp = np.linspace(0, 10, 1000)
Sb_high = np.log( np.sqrt(2 * np.pi * m * kB * temp) * L / h)

plt.figure(figsize=(10, 6))
plt.plot(temp, Sb_high, color='black')
plt.xlabel(r'T')
plt.ylabel(r'$S/k_B$')
plt.grid(True)
plt.show()
```

```python
temp = np.linspace(0, 10, 1000)
beta = 1 / (kB * temp)

Sb_low = np.log(np.exp(-beta * h ** 2 / 8 / m / L ** 2) + np.exp(-beta * 4 * h ** 2 / 8 / m / L ** 2))

plt.figure(figsize=(10, 6))
plt.plot(temp, Sb_low, color='black')
plt.xlabel(r'T')
plt.ylabel(r'$S/k_B$')
plt.grid(True)
plt.show()
```

```python
dL = 22 # nm
dE = 312.4 # pN nm
kBT = 4.114 # pN nm

f = np.linspace(9, 20, 1000)
r = 1 / (1 + np.exp(1 / kBT * (dE - dL * f)))

plt.figure(figsize=(10, 6))
plt.plot(f, r, 'b-')
plt.xlabel('Force (pN)')
plt.ylabel('Fraction Unfolded')
plt.grid(True)

# # Add text box with equation
# equation = r'$r = \frac{1}{1 + e^{\frac{1}{k_BT}(\Delta E - \Delta L \cdot f)}}$'
# plt.text(0.05, 0.95, equation, transform=plt.gca().transAxes,
#          verticalalignment='top', bbox=dict(boxstyle='round', facecolor='white', alpha=0.7))

plt.show()
```
